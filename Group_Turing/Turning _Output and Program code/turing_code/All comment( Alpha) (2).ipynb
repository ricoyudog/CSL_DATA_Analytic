```python
import json
import csv
import pandas as pd
import urllib.request
import math

x=0
y=0
Product_code=[]
date=[]
rating=[]
translate_content=[]
package_name=[]
product_id=[]
activites = pd.read_csv('data good url2.csv')
kkk=len(activites)
while(x<kkk):#change x = number of product should be all data
    with urllib.request.urlopen('https://www.klook.com/xos_api/v1/usrcsrv/activities/'+str(activites['Product_code'].iloc[x])+'/reviews?page=1&limit=10&translate=1&_=1570534320846') as url:
        source = json.loads(url.read().decode())
        total=source['result']['total']
        limit=source['result']['limit']
        page=total/limit+1
        print(activites['Product_code'].iloc[x])
        if page == 1:
            Product_code.append(str(activites['Product_code'].iloc[x]))
            date.append("No comment")
            rating.append("No comment")
            translate_content.append("No comment")
            package_name.append("No comment")
        else:
            i=1
            while(i<math.ceil(page)):
                with urllib.request.urlopen('https://www.klook.com/xos_api/v1/usrcsrv/activities/'+str(activites['Product_code'].iloc[x])+'/reviews?page='+str(i)+'&limit=10&translate=1&_=1570534320846') as url:
                    data = json.loads(url.read().decode())
                try:
                    for item in data['result']['item']:
                        Product_code.append(str(activites['Product_code'].iloc[x]))
                        date.append(item['date'])
                        rating.append(item['rating'])
                        translate_content.append(item['translate_content'])
                        package_name.append(item['package_name'])
                except:
                    Product_code.append(str(activites['Product_code'].iloc[x]))
                    date.append("NA")
                    rating.append("NA")
                    translate_content.append("NA")
                    package_name.append("NA")
                i+=1
        x+=1

   
    
df=pd.DataFrame({'Product_code':(Product_code),
                 'date':(date),
                    'rating':(rating),
                    'translate_content':(translate_content),
                'package_name':(package_name)})
```
