```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('commen all.csv',encoding='utf-8')
data.head()
```

```python
data['rating'].unique()
```

```python
def make_label(rating):
    if rating > 60:
        return 1
    else:
        return 0
    
data['sentiment'] = data.rating.apply(make_label)
```

```python
data.head()
```

```python
from snownlp import SnowNLP
```

```python
def snow_result(translate_content):
    s = SnowNLP(translate_content)
    if s.sentiments >= 0.6:
        return 1
    else:
        return 0
    
data['snlp_result'] = data.translate_content.apply(snow_result)
```

```python
counts = 0
for i in range(len(data)):
    if data.iloc[i,2] == data.iloc[i,3]:
        counts+=1

print(counts/len(data))
```

```python
data.head()
```

```python
import jieba

def chinese_word_cut(mytext):
    return " ".join(jieba.cut(mytext))

data['cut_comment'] = data.translate_content.apply(chinese_word_cut)
```

```python
data.head()
```

```python
X = data['cut_comment']
y = data.sentiment

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=22)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_custom_stopwords(stop_words_file):
    with open(stop_words_file) as f:
        stopwords = f.read()
    stopwords_list = stopwords.split('\n')
    custom_stopwords_list = [i for i in stopwords_list]
    return custom_stopwords_list

stop_words_file = 'stopword.txt'
stopwords = get_custom_stopwords(stop_words_file)

vect = CountVectorizer(max_df = 0.8, 
                       min_df = 3, 
                       token_pattern=u'(?u)\\b[^\\d\\W]\\w+\\b', 
                       stop_words=frozenset(stopwords))
```

```python
test = pd.DataFrame(vect.fit_transform(X_train).toarray(), columns=vect.get_feature_names())
test.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

X_train_vect = vect.fit_transform(X_train)
nb.fit(X_train_vect, y_train)
train_score = nb.score(X_train_vect, y_train)
print(train_score)
```

```python
X_test_vect = vect.transform(X_test)
print(nb.score(X_test_vect, y_test))
```

```python
X_vec = vect.transform(X)
nb_result = nb.predict(X_vec)
data['nb_result'] = nb_result
```
